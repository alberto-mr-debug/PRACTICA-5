En esta primera seccion tenemos la importacion de las librerias que vamos a utilizar, en este caso cv2 y matplotlib

```python
import cv2
import matplotlib.pyplot as plt
```

Esta parte nos servira para la carga de la imagen como para la conversion de esta en diferentes formatos

```python
# Cargar imagen oscura
img_bgr = cv2.imread('/content/IMAGEN 1.jpg')
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# Convertir a HSV
img_hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)

# Separar canales H, S y V
h, s, v = cv2.split(img_hsv)

# Ecualizar solo el canal de valor (brillo)
v_eq = cv2.equalizeHist(v)

# Combinar canales con el V ecualizado
img_hsv_eq = cv2.merge([h, s, v_eq])

# Convertir de nuevo a BGR y luego a RGB para visualización
img_eq_bgr = cv2.cvtColor(img_hsv_eq, cv2.COLOR_HSV2BGR)
img_eq_rgb = cv2.cvtColor(img_eq_bgr, cv2.COLOR_BGR2RGB)
```

esta ultima parte es mas que nada para dezplegar lo que es la imagen en comparacion de la original y por la que paso por varios filtros, como resultados tenemos cierta mejoria, claro en lo personal parece que la imagen se ve mas oscura que la anterior pero se pueden ver otro tipo de detalles, estos pasos son los mismos que con los otros dos ejemplos dejandonos ver una claridad diferente con respecto a detalles.

```python
# Mostrar imágenes original y mejorada
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title('Imagen original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img_eq_rgb)
plt.title('Imagen mejorada (Ecualización V)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
